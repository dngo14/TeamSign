```python
from PIL import Image
import numpy as np
import csv

import re

import pandas as pd
import matplotlib.pyplot as plt

def convert_back(letter, num): ## The function lets you see the converted csv file as the image again, and saves it as a png file
    df = pd.read_csv('./ASL_CSV/train/'+ letter + '/' + letter + str(num) + '.csv', sep=',', header=None)

    array = df.to_numpy()
    result = array.flatten()
    matrix = []

    for x in result:
        replaced = re.sub("[^0-9 ]+", "", x)
        replaced = re.sub(' +', ' ', replaced).strip()
        num_list = replaced.split(" ")
        num_list = [eval(i) for i in num_list]
        matrix.append(num_list)
        
    matrix = np.array(matrix)

    data = Image.fromarray(matrix)
    data.show()

    data = data.convert('RGB')
      
    # saving the final output as a PNG file
    data.save('./test' + letter + str(num) + '.png')

#convert_back('O', 2)


def convert_to_csv(letter): ## this function will convert the images for a letter given to the csv files
    for i in range(1,2):
        img = Image.open('./ASL_64/train/' + letter + '/' + letter + str(i) + '.jpg')
        imgGray = img.convert('L')

        px = imgGray.load()

        pixels = list(imgGray.getdata())
        
        width, height = imgGray.size
        pixels = [pixels[i * width:(i + 1) * width] for i in range(height)]

        pixels = np.array(pixels)

        #data = Image.fromarray(pixels)
        #data.show()

        f = open('./ASL_64/train/' + letter + '/'  + letter + str(i) + '.csv', 'w')

        # create the csv writer
        writer = csv.writer(f)

        # write a row to the csv file
        writer.writerow(pixels)

        # close the file
        f.close()

for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
    convert_to_csv(letter)

```
